# cv2. erode(src, kernel)
### 概要
攻撃力が半分になる速攻魔法を想起させる関数

モロフォロジー変換で使われる収縮処理を行う関数です

### 要点

・**モロフォロジー変換**という、二値画像に対して、収縮・膨張を繰り返してノイズを取り除く手法の**収縮(erosion)**で使われるのが、このerode()

・収縮処理とは、**画像の背景または穴に接する対象の画素をひとまわり剥ぎ取る処理**のこと

・この関数単体で使われることはあまりなく、delate()とセットで使うことが多い、



### 引数(公式リファレンスから引用)
**src**	        入力画像

**kernel**     	収縮で使うカーネルサイズ。すなわち、削り取る画素の大きさ。このカーネルが大きければ大きいほど、一回の収縮処理で削り取る画素が大きくなる。

**iteration**	 収縮処理を行う回数




### 参考URL
https://docs.opencv.org/master/d4/d86/group__imgproc__filter.html#gaabe8c836e97159a9193fb0b11ac52cf1

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
```

```python
img  = cv2.imread("../img/erode.png")
plt.rcParams['figure.figsize'] = (15 ,15)  #pltのサイズを変更
plt.subplots_adjust(wspace=0.4, hspace=0.6)

vals = (val_start,val_end,val_step) = 3,31,2#カーネルサイズが31まで
vals2 = (val_start,val_end,val_step) = 1,10,1 #イテレーション回数


@interact(kernelSize = vals,iteration = vals2)
def main(kernelSize =3,iteration = 1):
    kernel = np.ones((kernelSize, kernelSize), np.uint8)
    
    output_img = cv2.erode(img, kernel, iterations = iteration)
    
    #元画像
    plt.subplot(1,2,1)
    plt.title("origin") 
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    #ラプラシアンフィルタを当てた画像
    plt.subplot(1,2,2)
    plt.title("erode")
    plt.imshow(cv2.cvtColor(output_img, cv2.COLOR_BGR2RGB))
    
    
    plt.show()
```
