# getGaussianKernel（ksize、sigma）
### 概要

cv2.filter2D()が使いたいあなたに!!

ガウシアンフィルタを作成する関数です。

### 要点

・ガウスフィルタは**注目画素との距離に応じて**各画素ごとに重みが決める(平均値フィルタが全ての画素の重みが同一)平滑化フィルタのこと

・cv2.GaussianBlur()とgetGaussianKernel()の違いは、cv2.GaussianBlur()はガウシアンフィルタをかけた結果の画像を戻り値とするのに対して、この関数はガウシアンカーネルを戻り値として返す

・cv2.filter2D()を使う時など、カーネルが必要なときに使用する

### 引数(公式リファレンスから引用)


**ksize**     フィルタのサイズ。正方形のフィルタとなる。奇数で1より大きい数をいれる必要がある

**sigma**    ガウスカーネル標準偏差


### 戻り値
**retval**    ガウシアンフィルタ	     

### 参考URL
https://docs.opencv.org/master/d4/d86/group__imgproc__filter.html#gaabe8c836e97159a9193fb0b11ac52cf1

```python

import cv2
import matplotlib.pyplot as plt
from ipywidgets import interact
```

```python
img  = cv2.imread("../img/tower.png")
plt.rcParams['figure.figsize'] = (15 ,15)  #pltのサイズを変更
plt.subplots_adjust(wspace=0.4, hspace=0.6)

vals = (val_start,val_end,val_step) = 1,101,2
vals2 = (val_start,val_end,val_step) = 1,100,1
default = 5 
defaultSigma = 1

@interact(ksize = vals,sigma = vals2)
def main(ksize = default,sigma = defaultSigma):
    
   # ガウシアンカーネル
    gaussianK = cv2.getGaussianKernel(ksize,sigma)
    
    # 畳み込み
    img_blar = cv2.filter2D(img, -1,gaussianK)
    
    
    
    #元画像
    plt.subplot(1,2,1)
    plt.title("origin") 
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    #出力画像
    plt.subplot(1,2,2)
    plt.title("gaussian filter")
    plt.imshow(cv2.cvtColor(img_blar, cv2.COLOR_BGR2RGB))
    
    
    plt.show()
```
