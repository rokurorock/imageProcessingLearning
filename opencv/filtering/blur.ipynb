# cv2.blur(img,ax,ay)
#### 概要
平均値フィルタを用いて画像を平滑化する関数。

とにかく画像をボカしたい　そんなあなたにオススメ



#### 引数
img・・入力画像

ax ・・フィルタの横幅サイズ

ay  ・・フィルタの縦幅サイズ

```python
import cv2
import matplotlib.pyplot as plt
from ipywidgets import interact
```

```python
img  = cv2.imread("../../img/tower.jpg")
plt.figure(figsize = (30,5))

vals = (val_start,val_end,val_step) = 1,201,2
default = 5

@interact(N = vals)
def main(N = default):
    
    plt.subplots_adjust(wspace=0.4, hspace=0.6)
    img_blar = cv2.blur(img, (N,N))#平均値フィルタ
    
    #元画像
    plt.subplot(1,2,1)
    plt.title("origin") 
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    #ぼかした画像
    plt.subplot(1,2,2)
    plt.title("average filter")
    plt.imshow(cv2.cvtColor(img_blar, cv2.COLOR_BGR2RGB))
    
    
    plt.show()
```
