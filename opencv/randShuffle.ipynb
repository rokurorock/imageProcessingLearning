# 	cv2.randShuffle(dst [、iterFactor])
### 概要
なぜか引数に渡した画像もシャッフルされる不思議な関数

配列の中身をランダムでシャッフル関数
### 要点

・第一引数にnumpy配列を入れると

　"Expected Ptr<cv::UMat> for argument 'dst'"
   とエラーが出る注意
   
 ・Ptr<cv::UMat>　すなわち　cv2,imread()で読み込んだ画像を引数にしなければならない

・**第一引数dstに対して配列のランダムシャッフル**が実行される 

・ノイズのような画素がランダムに散りばめられた画像が出力される

### 引数(公式リファレンスから引用)
**dst**	opencv形式の配列(cv2,imread()で出力された配列)。関数の実行後引数である、dstの画像もランダムになることに注意
**iterFactor**	デフォルトは1 入れ替え処理の回数 (配列の列* 配列の行*iterFactorらしい


### 戻り値
**dst** ランダムに配列が並び替えられた画像(入力画像と同値)
 
### 参考URL
[公式ドキュメント](https://docs.opencv.org/master/d2/de8/group__core__array.html#ga6a789c8a5cb56c6dd62506179808f763)


```python
import cv2
import matplotlib.pyplot as plt
from ipywidgets import interact
import numpy as np
```

```python
def main():
    plt.rcParams['figure.figsize'] = (15 ,15)  #pltのサイズを変更
    plt.subplots_adjust(wspace=0.4, hspace=0.6)
    img  = cv2.imread("../img/rena.png")
    img2 = img.copy()
    img2 = cv2.randShuffle(img2)
    print("シャッフル前",img.shape)
    print("シャッフル後",img2.shape)
   
   
    cv2.imwrite("../img/test.png",img2)
    #元画像
    plt.subplot(1,2,1)
    plt.title("origin") 
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    #
    plt.subplot(1,2,2)
    plt.title("randam")
    plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
    
main()
```
