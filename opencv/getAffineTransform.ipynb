# cv2.cv.getAffineTransform(src,dst)

### 概要
任意のアフィン変換のための変換行列を作成する関数です!

### 要点
・アフィン変換は任意の平行四辺形から別の任意の平行四辺形への変換のため、平行四辺形は3点が決定すれば残りの1点も決まるので、変換前の3点の座標と変換後の3点の座標が与えられればアフィン変換の変換行列は一意に決まる。

・そのため、引数は変換前の3点の座標と変換後の3点の座標となる

### 引数(公式リファレンスから引用)
**src**	        変換前の3点の座標。

**dst**         変換後の3点の座標



### 戻り値
**mat**	      変換行列
### 参考URL
[公式ドキュメント](https://docs.opencv.org/4.0.1/da/d54/group__imgproc__transform.html#ga8f6d378f9f8eebb5cb55cd3ae295a999)

[Python, OpenCVで幾何変換アフィン変換・射影変換など](https://note.nkmk.me/python-opencv-warp-affine-perspective/)

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
#from ipywidgets import interact
```

```python
img  = cv2.imread("../img/girl.jpg") 
plt.rcParams['figure.figsize'] = (20 ,20)  #pltのサイズを変更
plt.subplots_adjust(wspace=0.3, hspace=0.2)


def main():
    img2 =  img.copy()
    src_pts = np.array([[30, 30], [50, 200], [350, 50]], dtype=np.float32)#変換前の三点
    dst_pts = np.array([[90, 20], [140, 170], [280, 80]], dtype=np.float32)#変換後の3点

    mat = cv2.getAffineTransform(src_pts, dst_pts)#変換行列を作成
    
    img2 = cv2.warpAffine(img2, mat, (int(img2.shape[1]/1), int(img2.shape[0]/1)))#バイオリニア補間
     
        

   
    #元画像
    plt.subplot(1,2,1)
    plt.title("origin") 
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    
    #バイリニア補間でアフィン変換
    plt.subplot(1,2,2)
    plt.title("bilinear") 
    plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
    
    
    plt.show()
main()
```
