# cv2.filter2D(src、ddepth、kernel)
### 概要
自家製カーネルが使えるのはこの関数だけ!!

任意のカーネルを用いて画像に対してフィルタリングをする関数です!

### 要点

・第三引数にnumpy配列で作成したカーネルを代入することで、任意のカーネルを使ったフィルタリングができる

・人によっては、フィルタリングはこの関数のみで行う人も多い(cv2.Laplacianなどは使わないでこの関数のみでフィルタリングを行う)

・今回の例で作るカーネルは**ソーベルフィルタ**。ソーベルフィルタは平滑化(ノイズを除去)した後にエッジを抽出することができる１次微分フィルタ。カーネルの中心の上下の重みを少し大きくしていることに、同様の1次微分フィルタであるプレフィットフィルタよりもエッジを検出しやすい 


### 引数(公式リファレンスから引用)
**src**	        入力画像

**ddepth**     	出力画像のビット深度　
[出力深度一覧](https://docs.opencv.org/master/d4/d86/group__imgproc__filter.html#filter_depths)
                      

**kernel**	   フィルタとして利用するカーネル。numpy配列で作成する必要がある




### 参考URL
[公式ドキュメント](https://docs.opencv.org/master/d4/d86/group__imgproc__filter.html#gaabe8c836e97159a9193fb0b11ac52cf1)

```python
import cv2
import matplotlib.pyplot as plt
from ipywidgets import interact
import numpy as np
```

```python
img  = cv2.imread("../img/tower.jpg")
plt.rcParams['figure.figsize'] = (15 ,15)  #pltのサイズを変更
plt.subplots_adjust(wspace=0.4, hspace=0.6)


def main():
    kernel = np.array([[1, 0, -1], #ソーベルフィルタ横
                     [2, 0, -2],
                     [1, 0, -1]])
    
    output_img = dx = cv2.filter2D(img, -1, kernel)
    
    #元画像
    plt.subplot(1,2,1)
    plt.title("origin") 
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    #ソーベルフィルタを当てた画像
    plt.subplot(1,2,2)
    plt.title("sobelFileter")
    plt.imshow(cv2.cvtColor(output_img, cv2.COLOR_BGR2RGB))
    
    
    plt.show()
main()
```
