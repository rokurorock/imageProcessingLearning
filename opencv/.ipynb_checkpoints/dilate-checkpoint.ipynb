# cv2. dilate(src, kernel)
### 概要
Go for it!! go for it!! 愛情こそが膨張ッ!!

モロフォロジー変換で使われる膨張処理を行う関数

### 要点

・**モロフォロジー変換**という、二値画像に対して、収縮・膨張を繰り返してノイズを取り除く手法の中の**膨張(dilation)**で使われるのがこのdilate()関数

・膨張処理とは、**対象の画素をひとまわり大きくする処理**のこと。つまり、erotionの逆となる

・この関数単体で使われることはあまりなく、erote()とセットで使うことが多い、



### 引数(公式リファレンスから引用)
**src**	        入力画像

**kernel**     	膨張で使うカーネルサイズ。すなわち、一度の膨張処理でどれくらい大きくなるかを決める

**iteration**	 膨張処理を行う回数

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
```

```python
img  = cv2.imread("../img/dilate.png")
plt.rcParams['figure.figsize'] = (15 ,15)  #pltのサイズを変更
plt.subplots_adjust(wspace=0.4, hspace=0.6)

vals = (val_start,val_end,val_step) = 3,31,2#カーネルサイズが31まで
vals2 = (val_start,val_end,val_step) = 1,10,1 #イテレーション回数


@interact(kernelSize = vals,iteration = vals2)
def main(kernelSize =3,iteration = 1):
    kernel = np.ones((kernelSize, kernelSize), np.uint8)
    
    output_img = cv2.dilate(img, kernel, iterations = iteration)
    
    #元画像
    plt.subplot(1,2,1)
    plt.title("origin") 
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    #ラプラシアンフィルタを当てた画像
    plt.subplot(1,2,2)
    plt.title("dilate")
    plt.imshow(cv2.cvtColor(output_img, cv2.COLOR_BGR2RGB))
    
    
    plt.show()
```
