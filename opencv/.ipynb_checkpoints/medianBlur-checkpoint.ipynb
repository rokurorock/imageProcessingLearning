# cv2.medianBlur(src, ksize)
### 概要
中央値フィルタを用いて空間フィルタリングする関数。
ノイズは消したいけど、画像は平滑化したくないときに

### 要点

・**メディアンフィルタ**は画像のノイズを除去するフィルタの一つで、注目画素の周囲の**中央値**を出力するフィルタである

・メディアンフィルタは画像のエッジを保ちつつ、ノイズなどを削除することができる


### 引数(公式リファレンスから引用)
**src**	        入力画像

**ksize**     フィルタのサイズ。正方形のフィルタとなる。奇数で1より大きい数をいれる必要がある
https://docs.opencv.org/master/d2/de8/group__core__array.html#ga209f2f4869e304c82d07739337eae7c5


### 参考URL
https://docs.opencv.org/master/d4/d86/group__imgproc__filter.html#gaabe8c836e97159a9193fb0b11ac52cf1

```python
import cv2
import matplotlib.pyplot as plt
from ipywidgets import interact
```

```python
img  = cv2.imread("../img/tower.png")
plt.figure(figsize = (50,10))

vals = (val_start,val_end,val_step) = 1,201,2
default = 5 
defaultSigma = 1

@interact(ksize = vals)
def main(ksize = default):
    
    plt.subplots_adjust(wspace=0.4, hspace=0.6)
    img_blar =cv2.medianBlur(img, ksize)#ガウシアンフィルタ
    
    #元画像
    plt.subplot(2,1,1)
    plt.title("origin") 
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    #ぼかした画像
    plt.subplot(2,1,2)
    plt.title("median filter")
    plt.imshow(cv2.cvtColor(img_blar, cv2.COLOR_BGR2RGB))
    
    
    plt.show()
```
